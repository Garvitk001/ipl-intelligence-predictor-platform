```python
import pandas as pd
import os

print("Starting Live In-Game Exhaustion Baseline Engineering...")

# 1. Load Data
deliveries = pd.read_csv('../data/raw/deliveries.csv')

# 2. Batter Exhaustion (Cumulative balls faced in the match so far)
# We group by match_id and batter, then use cumcount to find how many balls they've faced up to this exact delivery
deliveries['current_batter_balls_faced'] = deliveries.groupby(['match_id', 'batter']).cumcount() + 1

# 3. Bowler Exhaustion (Cumulative balls bowled in the match so far)
deliveries['current_bowler_balls_bowled'] = deliveries.groupby(['match_id', 'bowler']).cumcount() + 1

# 4. Create the Baseline Exhaustion Metric
# Fast bowling is incredibly taxing, so we weight bowler balls (1.5) higher than batter balls (0.5)
# Note: In Phase 6, our Streamlit app will pull live Weather API data and multiply it against this base score!
deliveries['base_exhaustion_score'] = (deliveries['current_batter_balls_faced'] * 0.5) + (deliveries['current_bowler_balls_bowled'] * 1.5)

# 5. Extract just the features we need
exhaustion_df = deliveries[['match_id', 'inning', 'over', 'ball', 'batter', 'bowler', 
                            'current_batter_balls_faced', 'current_bowler_balls_bowled', 'base_exhaustion_score']]

# 6. Save the Intelligence Data
os.makedirs('../data/processed', exist_ok=True)
exhaustion_df.to_csv('../data/processed/in_game_exhaustion.csv', index=False)

print("✅ Live In-Game Exhaustion baseline calculated successfully!")
print("\nSample: Extreme Exhaustion Scenarios (Deep into the innings with set batters and tired bowlers):")

# Display a sample of deliveries deep in the match where the batter has faced 40+ balls 
display(exhaustion_df[(exhaustion_df['over'] > 15) & (exhaustion_df['current_batter_balls_faced'] > 40)].head(10))
```
